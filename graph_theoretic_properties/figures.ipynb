```python
import matplotlib.pyplot as plt
import numpy as np

# Generate points to plot
x = np.linspace(0,8,200)
y = 2/(1 + np.exp(x))

# Plotting x and y
fig = plt.figure(figsize=(4,4), dpi=300)
plt.gca().grid(True)
ax = fig.add_subplot(1, 1, 1)

# Move left y-axis and bottom x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x, y, 'purple')
plt.plot(0,1,'ro')

# Define custom tick positions and labels
_lim = 6
x_ticks = list(range(-_lim,_lim+1))
x_ticks.remove(0) # remove the 0
y_ticks = list(range(-_lim,_lim+1))
ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks)

# Manually set the tick labels, removing one of the 0's
x_tick_labels = [str(x) for x in x_ticks]
y_tick_labels = [str(y) for y in y_ticks]
ax.set_xticklabels(x_tick_labels, fontsize=6)
ax.set_yticklabels(y_tick_labels, fontsize=6)

plt.xlim([-_lim, _lim])
plt.ylim([-_lim, _lim])
plt.title('Plot of function used Coherence',fontsize=8)
plt.savefig(fname= "D:\D_Drive\BITS_Pilani\Year_5(2023-24)\\thesis_plots\coherence_formula_plt")
plt.show()
```
