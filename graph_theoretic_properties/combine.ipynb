```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("syn1.csv", usecols=["Order","Size","Average Degree","Diameter","Sparsity","Coherence"])
df2 = pd.read_csv("syn2.csv", usecols=["Order","Size","Average Degree","Diameter","Sparsity","Coherence"])
df5 = pd.read_csv("syn5.csv", usecols=["Order","Size","Average Degree","Diameter","Sparsity","Coherence"])
```

```python
# Plot Sparsity
x1 = list(range(len(df1)))
x2 = list(range(len(df2)))
x5 = list(range(len(df5)))

sparse1 = list(df1['Sparsity'])
sparse2 = list(df2['Sparsity'])
sparse5 = list(df5['Sparsity'])

# mean_sparse = list(np.ones(len(df))*df['Sparsity'].mean())
plt.figure(figsize=(20,6), dpi = 500)
plt.gca().grid(True)
# plt.gca().set_facecolor((1, 1, 0.8))
plt.plot(x1, sparse1, label='syn1 sparsity',color='blue')
plt.plot(x2, sparse2, label='syn2 sparsity',color='red')
plt.plot(x5, sparse5, label='syn5 sparsity',color='k')
# plt.plot(x,def_sparse, label = 'Max edge density for sparse graphs')
# plt.plot(x,mean_sparse, label = 'Average sparsity', color='k')

# sp_mean = df['Sparsity'].mean()
# sp_std = df['Sparsity'].std()
# y1 = sp_mean - sp_std
# y_1 = y1 - sp_std
# y2 = sp_mean + sp_std
# y_2 = y2 + sp_std
# plt.axhspan(y_1, y_2, color='red', alpha=0.3)
# plt.axhspan(y1, y2, color='green', alpha=0.3)

plt.title('Sparsity of explanations by GNNExplainer for all datasets', fontsize=20)
plt.legend(loc="lower right",bbox_to_anchor=(0.85, 0.06),fontsize=15)
plt.ylabel("Sparsity",fontsize=15)
plt.xlabel("Index",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.savefig(fname= "D:\D_Drive\BITS_Pilani\Year_5(2023-24)\\thesis_plots\GNNExp\GNN_sparsity_datasets")
plt.show()
```

```python
# Plot Coherence
x1 = list(range(len(df1)))
x2 = list(range(len(df2)))
x5 = list(range(len(df5)))

coh1 = list(df1['Coherence'])
coh2 = list(df2['Coherence'])
coh5 = list(df5['Coherence'])

# mean_sparse = list(np.ones(len(df))*df['Sparsity'].mean())
plt.figure(figsize=(20,6), dpi = 500)
plt.gca().grid(True)
# plt.gca().set_facecolor((1, 1, 0.8))
plt.plot(x1, coh1, label='syn1 coherence',color='blue')
plt.plot(x2, coh2, label='syn2 coherence',color='red')
plt.plot(x5, coh5, label='syn5 coherence',color='k')
# plt.plot(x,def_sparse, label = 'Max edge density for sparse graphs')
# plt.plot(x,mean_sparse, label = 'Average sparsity', color='k')

# sp_mean = df['Sparsity'].mean()
# sp_std = df['Sparsity'].std()
# y1 = sp_mean - sp_std
# y_1 = y1 - sp_std
# y2 = sp_mean + sp_std
# y_2 = y2 + sp_std
# plt.axhspan(y_1, y_2, color='red', alpha=0.3)
# plt.axhspan(y1, y2, color='green', alpha=0.3)

plt.title('Coherence of explanations by GNNExplainer for all datasets', fontsize=20)
plt.legend(loc="upper right",fontsize=15)
plt.ylabel("Sparsity",fontsize=15)
plt.xlabel("Index",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.savefig(fname= "D:\D_Drive\BITS_Pilani\Year_5(2023-24)\\thesis_plots\GNNExp\GNN_coherence_datasets")
plt.show()
```
